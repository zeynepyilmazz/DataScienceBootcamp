**(1)** Create a blank figure with matplotlib functions. Plot cosine function for the numbers beetwen -3 π  and 3 π  with 0.1 intervals.

```
import matplotlib.pyplot as plt
import numpy as np
import math
```

```
x = np.arange(-3*math.pi ,3*math.pi,0.1)
y = np.cos(x)
fig , ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

**(2)** This time, create the same plot in red color with object methods.

```
x = np.arange(-3*math.pi ,3*math.pi,0.1)
y = np.cos(x)
fig , ax = plt.subplots()
ax.plot(x,y,color = '#B22222')
plt.show()
```
