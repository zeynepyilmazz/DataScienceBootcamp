In this assignment you will continue to make some plots on the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data). For plotting you will use Seaborn library. 

**(1)** Plot a line plot with seaborn for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
df.head()
```

```python
covid = df[df['location'].isin(['Spain', 'France', 'Germany','Italy'])]
covid.head()
```

```python
plt.figure(figsize=(10, 6))
plt.title('TOTAL DEATHS')
sns.lineplot(x = 'date' , y = 'total_deaths', data = covid[covid['date'] > '2020-04-01'], hue = 'location')
plt.xlabel('DATES')
plt.ylabel('COUNT OF DEATHS')
plt.show()
```

**(2)** Plot a bar plot with seaborn for average death number that compares continents.

```python
plt.figure(figsize=(9, 6))
sns.barplot(x = 'continent' , y = 'total_deaths' , data = df.groupby(by='location')[['continent','total_deaths']].last())
plt.xlabel('CONTÄ°NENTS')
plt.ylabel('AVERAGE COUNT OF DEATHS')
plt.xticks(rotation = 90)
plt.show()
```

**(3)**  Plot a histogram for daily deaths for any country you choose. Make four subplots for different bin counts and `kde` arguments.

```python
plt.figure(figsize=(15, 15))
plt.suptitle('GERMANY' , size=10 , color = 'blue')
plt.subplot(2,2,1)
sns.distplot(df[df['location'] == 'Germany']['new_cases'].dropna() , bins=50 , kde = False , color = '#FF914D')
plt.title("NEW DEATHS\n" + "bins : 50 / kde : False")
plt.xlabel('NEW DEATHS')
plt.ylabel('COUNT')
plt.subplot(2,2,2)
sns.distplot(df[df['location'] == 'Germany']['new_cases'].dropna(), bins=20 , kde = False , color = '#FF914D')
plt.title("NEW DEATHS\n" + "bins : 20 / kde : False")
plt.xlabel('NEW DEATHS')
plt.ylabel('COUNT')
plt.subplot(2,2,3)
sns.distplot(df[df['location'] == 'Germany']['new_cases'].dropna() , bins=50 , color = '#FF914D')
plt.title("NEW DEATHS\n" + "bins : 50 / kde : True")
plt.xlabel('NEW DEATHS')
plt.ylabel('NORMALIZED COUNT')
plt.subplot(2,2,4)
sns.distplot(df[df['location'] == 'Germany']['new_cases'].dropna() , bins=20 , color = '#FF914D')
plt.title("NEW DEATHS\n" + "bins : 20 / kde : True")
plt.xlabel('NEW DEATHS')
plt.ylabel('NORMALIZED COUNT')
plt.show()
```

**(4)** Create a figure and three subplots containing boxplot, violin plot and swarm plot for daily deaths of two countries you choose.

```python
plt.figure(figsize=(16,6))
plt.subplot(1,3,1)
plt.title('BOX PLOT')
sns.boxplot(x = 'location' ,  y = 'new_deaths', 
            data = df[df['location'].isin(['France' , 'Germany'])], palette = ['#FF914D', '#22B573'])
plt.xlabel('COUNTRY')
plt.ylabel('DAILY DEATHS')

plt.subplot(1,3,2)
plt.title('VIOLIN PLOT')
sns.violinplot(x = 'location',  y = 'new_deaths',
               data = df[df['location'].isin(['France' , 'Germany'])], palette = ['#FF914D', '#22B573'])
plt.xlabel('COUNTRY')
plt.ylabel('DAILY DEATHS')

plt.subplot(1,3,3)
plt.title('SWARM PLOT')
sns.swarmplot(x = 'location',  y = 'new_deaths',
               data = df[df['location'].isin(['France' , 'Germany'])], palette = ['#FF914D', '#22B573'])
plt.xlabel('COUNTRY')
plt.ylabel('DAILY DEATHS')

plt.show()
```
