In this assignment you will continue to make some plots on the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data). For plotting you will use Plotly Express. 

```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
df.head()
```

**(1)** Plot a line plot with plotly express for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
covid = df[(df['location'].isin(['Spain' , 'France' , 'Germany' , 'Italy'])) & (df['date'] > '2020-04-01')]
fig = px.line(covid, x = 'date', y = 'total_deaths', title='Total Deaths' , color = 'location')
fig.show()
```

**(2)** Plot a bar plot with plotly express for average death number that compares continents.

```python
covid_19 = df.groupby(by='location')[['continent' ,'total_deaths']].last()
covid_19 = covid_19.groupby(by='continent').mean()
fig = px.bar(data_frame = covid_19, 
             x= covid_19.index, 
             y='total_deaths' , color = covid_19.index)
fig.show()
```

**(3)** Plot a histogram for daily deaths for any country you choose. 

```python
fig = px.histogram(df[df['location'] == 'United States'], x="new_deaths"  )
fig.show()
```
